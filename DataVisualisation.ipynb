```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file=pd.read_csv("Heart.csv",usecols=['Age','RestingBP','Cholesterol','FastingBS','MaxHR','HeartDisease'])
file=file.head(40)
file
```

```python
file.plot()
```

```python
file.plot(subplots=True)
```

```python
file.plot(x='MaxHR',y='HeartDisease')
```

```python
file.plot(kind='hist',figsize=(50,10))
```

```python
file.plot(kind='bar',figsize=(50,10))
```

```python
file.plot(kind='box',y='Age')
```

```python
file.plot(kind='area',figsize=(50,10))
```

```python
file.plot(kind='density',figsize=(50,10))
```

```python
file.groupby(['Age']).sum().plot(kind='pie',y='HeartDisease',autopct='%1.0f%%',figsize=(50,10))
```

```python
file.groupby(['FastingBS']).sum().plot(kind='pie',y='HeartDisease',autopct='%1.0f%%')
```

```python
file.groupby(['Cholesterol']).sum().plot(kind='pie',y='HeartDisease',figsize=(50,10))
```
