```python
import pandas as pd
file=pd.read_csv("Heart.csv")
file
```

```python
file=pd.get_dummies(file,columns=['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope'])
f=file.pop('HeartDisease')
file.insert(len(list(file)),'HeartDisese',f)
file
```

```python
x=file.iloc[:,:-1].values
y=file.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(x_train,y_train)
```

```python
y_pred=gnb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
acc=accuracy_score(y_pred,y_test)
print('Accuracy Score = %.2f'%(acc*100))
prec=precision_score(y_pred,y_test)
print('Precision Score = %.2f'%(prec*100))
recall=recall_score(y_pred,y_test)
print('Recall Score = %.2f'%(recall*100))
f1=f1_score(y_pred,y_test)
print('F1 Score = %.2f'%(f1*100))
```

```python
from sklearn.naive_bayes import BernoulliNB
bnb=BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
y_pred=bnb.predict(x_test)
```

```python
acc=accuracy_score(y_pred,y_test)
print('Accuracy Score = %.2f'%(acc*100))
prec=precision_score(y_pred,y_test)
print('Precision Score = %.2f'%(prec*100))
recall=recall_score(y_pred,y_test)
print('Recall Score = %.2f'%(recall*100))
f1=f1_score(y_pred,y_test)
print('F1 Score = %.2f'%(f1*100))
```
